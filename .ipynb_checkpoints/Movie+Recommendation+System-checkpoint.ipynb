# Movie_Recommender_System

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# importing movie dataset
movie_dataset=pd.read_csv('F:\ML_DataSets\Self_projects\Movies data\movies.csv')
```

```python
movie_dataset.head()
```

```python
movie_dataset.tail()
```

```python
movie_dataset.columns
```

```python
# # selecting the required features
selected_features=['genres','keywords','tagline','cast','director','keywords','overview']
```

```python
print(selected_features)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(movie_dataset[selected_features].isnull())
```

```python
# fillingnull values with  empty string
movie_dataset[selected_features]=movie_dataset[selected_features].fillna('')
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(movie_dataset[selected_features].isnull())
```

```python
selected_features
```

```python
# # combinig textual data in one for converting in numerical form
# combined_feature=movie_dataset['genres']+ ' '+ movie_dataset['keywords']+ ' '+ movie_dataset['tagline']+' '+ movie_dataset['cast']+ ' ' + movie_dataset['director']
```

```python
combined_feature=' '
for i in selected_features:
    combined_feature+=  (movie_dataset[i]) + ' '
 
```

```python
combined_feature
```

```python
vec=TfidfVectorizer()
```

```python
feature_vector=vec.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python
# finding similarity between all movies
similarity=cosine_similarity(feature_vector)
print(similarity)
```

```python
similarity.shape
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(similarity)
```

# Combining all code for better  visualization

```python
# getting movie name
movie=input('Type your favourite movie name :')
print()

# creating a list containg all movie name
total_movie_list=movie_dataset['title'].tolist()

# finding the close match list for the movie name
close_movie_list=difflib.get_close_matches(movie,total_movie_list)

# getting closest movie name
close_movie=close_movie_list[0]

# finding closest movie name index
movie_index=movie_dataset[movie_dataset.title==close_movie]['index'].values[0]

# finding similarity movie list
similar_movie_list=list(enumerate(similarity[movie_index]))

# sortin movie based on similarity
sorted_similar_movie_list=sorted(similar_movie_list,key= lambda x: x[1] ,reverse=True)

# print recommended movies
i=1
print('Recommended Movies for you--->')
for m in sorted_similar_movie_list:
    name=movie_dataset[movie_dataset.index==m[0]]['title'].values[0]
    if(i<11):
        print( i,':', name)
        i+=1
```
